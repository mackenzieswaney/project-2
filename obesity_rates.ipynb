```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load
file = "National_Obesity_By_State.csv"

# Read Purchasing File and store into Pandas data frame
obesity_data= pd.read_csv(file)
obesity_data.head()
```

```python
obesity_rates_data = obesity_data[['OBJECTID', 'NAME', 'Obesity']]
obesity_rates_data.head()
```

```python
# Remove Puerto Rico & DC
good_obesity = obesity_rates_data.drop([38, 35])
good_obesity
```

```python
# Save data frame to CSV
good_obesity.to_csv('good_obesity.csv')

# Display data frame head 
good_obesity
```
